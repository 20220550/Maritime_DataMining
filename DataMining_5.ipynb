```python
## 20220550 윤명섭
## 해사 데이터마이닝 5주차 과제

import pandas as pd
import numpy as np

# github의 자동차 생산 자료를 읽어들인다
path = "https://github.com/dongupak/DataML/raw/main/csv/"
file = path+"vehicle_prod.csv"
df = pd.read_csv(file, index_col=0)

print(df, "\n")

# Total 행 추가
data1 = [df.sum(axis= 0)]

df2 = pd.DataFrame(data1,
                   columns= ["2007", "2008", "2009", "2010", "2011"],
                   index= ["Total"])

df3 = pd.concat([df, df2])
print("Total 행 추가 \n", df3, "\n")

# Average 행 추가
data2 = [df.mean(axis= 0)]

df4 = pd.DataFrame(data2,
                   columns= ["2007", "2008", "2009", "2010", "2011"],
                   index= ["Average"])

df5 = pd.concat([df3, df4])
print("Average 행 추가 \n", df5)
```
