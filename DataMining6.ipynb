```python
# xml 파일 다운로드
import kagglehub

# Download latest version
path = kagglehub.dataset_download("agileteam/bigdatacertificationkr")

print("Path to dataset files:", path)
```

```python
## xml 파일
## https://www.kaggle.com/datasets/agileteam/bigdatacertificationkr?select=basic1_xml.xml
## 에서 basic1_xml 파일을 선택했습니다.

import pandas as pd
import xml.etree.ElementTree as ET

# xml 파일 경로
xml_path = "/content/xmlData/basic1_xml.xml"

# xml 파일 읽기
tree = ET.parse(xml_path)
# 루트 요소 가져오기
root = tree.getroot()

# 특정 요소 목록 가져오기
elements = root.findall("row")

# DataFrame 생성
df = pd.DataFrame(columns=["city", "age", "f4"])

# 각 요소를 DataFrame에 추가
for element in elements:
  row = {
      "city": element.find("city").text,
      "age": float(element.find("age").text),
      "f4": element.find("f4").text
  }
  df2 = pd.DataFrame(row, index=[0])
  df = pd.concat([df, df2], ignore_index=True)

# DataFrame 정보 출력
print(df.info(), "\n")
# DataFrame 상위 5개 행 출력
print(df.head())
```
