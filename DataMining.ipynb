```python
## 20220550 윤명섭
## 해사 데이터마이닝 (1주차)과제1

## 해사 데이터마이닝 3주차 과제
# 1번 문제
print("1번 문제\n")
a = [2,3,4,5,6]
print("변경 전 리스트:", a)
revA = []

for _ in range(len(a)):
  revA.append(a.pop())

print(revA)

# 2-1번 문제
print("\n2번 문제\n")
tup = (1,2,5,4,3,2,1,4,7,8,9,9,3,7,3)
print("주어진 튜플:", tup)
temp1 = sorted(list(tup))

maxCount = 1
count = 1
curValue = maxValue = temp1[0]

for i in range(1, len(temp1)):
  if temp1[i] == curValue:
    count += 1
    if count >= maxCount:
      maxCount = count
      maxValue = curValue
  else:
    curValue = temp1[i]
    count = 1

print("가장 많이 나타내는 요소:", maxValue)

# 2-2번 문제
tup2 = (1,2,5,4,3,2,9,4,7,8,9,9,3,7,3)
print("주어진 튜플:", tup2)
temp2 = sorted(list(tup2))

maxCount = 1
count = 1
curValue = maxValue = temp2[0]

for i in range(1, len(temp2)):
  if temp2[i] == curValue:
    count += 1
    if count >= maxCount:
      maxCount = count
      maxValue = curValue
  else:
    curValue = temp2[i]
    count = 1

print("가장 많이 나타내는 요소:", maxValue)

# 3-1, 2번 문제
print("\n3번 문제\n")
menu1 = {'Americano':3000, 'Ice Americano':3500, 'Cappuccino':4000, 'Cafe Latte': 4500, 'Expresso':3600}

for key in menu1:
  print("{:16s} 가격: {:,}원".format(key, menu1[key]))

choice1 = input("위의 메뉴 중 하나를 선택하세요:")
if choice1 in menu1.keys():
  print("{}는 {:,}원입니다. 결제를 부탁합니다.".format(choice1, menu1[choice1]))
else:
  print("미안합니다. {}는 메뉴에 없습니다.".format(choice1))
```
